```python
df = df.rename(columns = {'Номер':'number', 
               'Дата создания':'create_date', 
               'Дата оплаты':'payment_date', 
               'Title':'title', 
               'Статус':'status',
               'Заработано':'money', 
               'Город':'city', 
               'Платежная система':'payment_system'})
```

```python
df.head()
```

```python
all_money = df.money.sum() #adding new var to further analysis 
```

```python
all_money
```

```python
money_by_city = df \
    .groupby(['title', 'city'], as_index = False)\
    .agg({'money':'sum'})\
    .sort_values('money', ascending = False)#groupping and aggregating
```

```python
money_by_city.to_csv('money_by_city.csv', index = False) #saving as csv
```

```python
money_title = df.query("status == 'Завершен'")\
    .groupby('title', as_index = False)\
    .agg({'money':'sum', 'number' : 'count'})\
    .sort_values('money', ascending = False) \
    .rename(columns = {'number':'success_orders'})#filtering
```

```python
today_day = datetime.today().strftime('%Y-%m-%d')
file_name = 'money_title_{}.csv' #scritping
```

```python
file_name = file_name.format(today_day)
```

```python
if int(money_title.money.sum())== int(all_money):
    print('File {} is written'.format(file_name))
    money_title.to_csv(file_name, index = False)
else:
    print('Error')
```

```python
taxi = pd.read_csv(r'C:\Users\aimbe\Downloads\Karpov\lesson_2_taxi_nyc.csv',encoding='windows-1251', sep = ',')
```

```python
taxi = taxi \
        .rename(columns ={'pcp 01':'pcp_01', 
                          'pcp 06':'pcp_06', 
                          'pcp 24':'pcp_24'})
```

```python
taxi
```

```python
min_pickups = taxi\
    .groupby('borough')\
    .agg({'pickups' : 'sum'})\
    .sort_values('pickups') \
    .idxmin()

#calculate min index
```

```python
taxi\
    .groupby(['borough', 'hday'])\
    .agg({'pickups' : 'mean'})\
    .sort_values('pickups') 
```

```python
pickups_by_mon_bor = taxi\
    .groupby(['pickup_month', 'borough'], as_index = False)\
    .agg({'pickups' : 'sum'})\
    .sort_values('pickups', ascending = False) 
```

```python
import pandas as pd 
def temp_to_celcius(temp):
    temp_c = []
    for i in temp:
        a = (((i-32)*5)/9)
        temp_c.append(a)
    temp_c = pd.Series(temp_c)
    return temp_c
```
