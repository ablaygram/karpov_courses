```python
def to_lower_underscore(nm):
  nm = nm.lower().replace(' ', '_')
  return nm
bookings=bookings.rename(columns=to_lower_underscore)
bookings
```

```python
bookings \
    .query('is_canceled != 0') \
    .groupby('country', as_index = False) \
    .agg({'is_canceled' : 'sum'}).sort_values('is_canceled', ascending = False).head(10) #top-10 country by succesful orders
```

```python
round(bookings\
    .query("hotel in ('Resort Hotel','City Hotel')")\
    .groupby('hotel', as_index = False)\
    .agg({'stays_total_nights' : 'mean'}),2) #mean of nights in hotel types slice
```

```python
bookings.query('reserved_room_type != assigned_room_type').shape #number of cases where assigned and reserved rooms are different
```

```python
bookings\
    .groupby(['arrival_date_year','arrival_date_month'], as_index = False)\
    .agg({'arrival_date_month' : 'count'}) #another analysis
```

```python
bookings\
    .query("hotel == 'City Hotel' and is_canceled == 1")\
    .groupby(['arrival_date_year','arrival_date_month'], as_index = False)['arrival_date_month'].size()

    #calculating amount of cancelation per monthes 
```

```python
bookings['adults'].mean() #easy computations
```

```python
bookings['total_kids'] = bookings['children'] + bookings['babies']  
```

```python
bookings\
    .groupby('hotel', as_index = False)\
    .agg({'total_kids':'mean'})
```

```python
bookings['has_kids'] = np.where(bookings['total_kids'] != 0, True, False)
```

```python
bookings\
    .query('is_canceled == 1')\
    .groupby('has_kids')\
    .agg({'has_kids': 'count'}) #some funny analysis
```
